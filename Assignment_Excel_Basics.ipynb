# Difference Between Relative, Absolute and Mixed Cell Referencing

1. Relative Cell Reference
  - A relative reference changes automatically when the formula is copied to another cell. It adjusts based on the new row or column. Example: If the formula in C1 is `=A1 + B1` and you copy it to `C2`, it becomes `=A2 + B2`.
2. Absolute Cell Reference
  - An absolute reference does not change when copied. It stays fixed because the `$` symbol locks the row and column. Example: `$A$1` remains `$A$1` even if copied anywhere in the sheet.
3. Mixed Cell Reference
  - A mixed reference has either the row or the column fixed. Two types: `$A1` (column fixed) and `A$1` (row fixed). Example: If `A$1` is copied down, row 1 stays the same but the column changes.
